So, here we have some linear regression model with predictors(TV, radio, newspaper) and response(Sales).
Ans here we see some statistics. So, first of all, lets remind to oursefles - what is null hypotesis? Here this is a test,
that consists of $H_0$ and $H_a$, where:
- $H_0$ - the hypotesis, that all coefficients of a predicors($B_1$, $B_2$, $B_3$) are all equal zero, which means that corresponding
predictors does not contain some useful info to our response: $B_1$ = $B_2$ = $B_3$ = 0
- $H_a$ - at least one predictor has non-zero coefficient.
This type of test is importat because before building some models, we want to know if our predictors has some influence on response.

In order to reject or apporove null hypotesis, we compute t-statistic to each coefficient:
$$t=\frac{\hat{\beta}_1-0}{\mathrm{SE}(\hat{\beta}_1)}$$
and then, we get some p-value, which helps us to understand if coeffcient is non-zero. In general, if p < 0.05, then coefficent is non-zero.

If we look at the table, we see that TV anf radio has very small p-value, while newspaper has quite huge - 0.8599.
So, we can make assumption, that in this model - newspapper is not important and does not carry some useful info since the corresponding
coefficient is approximately equal to zero.

2. Carefully explain the differences between the KNN classifier and KNN
regression methods
---

So, the main difference between KNN clasifier and KNN regression is that they have different goals.

KNN classifier: this method is used in CLASSIFICATION tasks. The main idea is to take some K(number of neighbors), and 
the class for every new test point will be selected based on number of classes in the neighborhood:
$$
\Pr(Y = j \mid X = x_0) = \frac{1}{K} \sum_{i \in N_0} I(y_i = j)
$$
,
where for each value in neigborhood we indicate whether it 1 - belongs to some class 1 or 0 if not, and in result we get the propotion -
the probability of new test point to be in class i

KNN regression: this method is used for REGRESSION problems. The concept us the same: we find some k nearest neighbors, sum them, and
their mean - new test point:
$$
\hat{f}(x_0)=\frac{1}{K}\sum_{x_i\in\mathcal{N}_0}y_i.
$$
